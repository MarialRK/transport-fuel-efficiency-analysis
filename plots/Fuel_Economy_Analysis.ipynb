```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('fuel_econ.csv')

# Initial exploration
print(df.head())  # First few rows
print(df.info())  # Column info and data types
print(df.describe())  # Statistical summary
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values if needed
# df = df.dropna() or fill with appropriate values
```

```python
# Plotting histogram for combined fuel efficiency (MPG)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['comb'], bins=20, kde=True, color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Combined MPG (Miles per Gallon)', fontsize=14)
plt.xlabel('Combined MPG', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('histogram_combined_mpg.png')  # Save for submission
plt.show()
```

# The histogram shows that most vehicles fall between 15 and 35 MPG, indicating typical fuel efficiency ranges.
# A smaller number of vehicles exceed 40 MPG, likely hybrid or electric vehicles.
# This insight is critical for sustainability-focused transport planning.
# To reduce emissions, promoting high-MPG vehicles should be a policy and design priority.


```python
# Correlation Heatmap
plt.figure(figsize=(14, 10))
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numeric Features', fontsize=16)
plt.tight_layout()
plt.savefig('heatmap_correlation_matrix.png')  # Save for submission
plt.show()
```

# Interpretation of Correlation Matrix:
# - 'city', 'highway', and 'comb' MPG are highly correlated (> 0.9). This makes sense because 'comb' is derived from the other two.
# - For AI modeling, one of 'city' or 'highway' could be dropped to reduce redundancy and multicollinearity.
# - 'co2' shows a strong negative correlation with MPG values, indicating better fuel economy equals lower CO2 output.
# - 'feScore' and 'ghgScore' are moderately correlated with MPG, useful for scoring vehicle environmental performance.
